We have converted important features into integers, and now we can handle the missing values

using simple impute applied mean stratergy on SuperOver and WinnigTeam Features

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan, strategy="mean")
dataset["SuperOver"]=si.fit_transform(dataset.SuperOver.values.reshape(-1,1))
dataset["WinningTeam"]=si.fit_transform(dataset.WinningTeam.values.reshape(-1,1))
```

apply dropna method on Player of match and city

```python
dataset=dataset.dropna(subset=["Player_of_Match"])
dataset=dataset.dropna(subset=["City"])
```

According to my logic, I can't perceive margin as important for the analysis, so I have decided to drop the Margin feature

```python
dataset.drop(["Margin"],axis=1,inplace=True)
```
